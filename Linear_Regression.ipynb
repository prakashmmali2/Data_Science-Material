```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (20.0,10.0)


```

```python
# Reading the Data

plt.rcParams['figure.figsize'] = (20.0,10.0)
data = pd.read_csv('https://github.com/RupeshMohan/Linear_Regression/raw/master/headbrain.csv')
print(data.shape)
data.head()
```

```python
x = data['Head Size(cm^3)'].values
y = data['Brain Weight(grams)'].values

mean_x = x.mean()
mean_y = y.mean()

print(mean_x)
print(mean_y)
```

```python
m_x = np.mean(x)
m_y = np.mean(y)

n = len(x)

num = 0
demon = 0

for i in range(n):
    num = num + (x[i] - m_x) * (y[i] - m_y)
    demon = demon + (x[i] - m_x) ** 2

b1 = num/demon
d0 = m_y - (b1 * mean_x)

print(b1 , d0)
```

```python
max_x = np.max(x) +  100
min_x = np.min(x) - 100

x = np.linspace(min_x,max_x,1000)
y = d0 + b1 * x


plt.plot(x , y , color = '#58b970' , label = 'Regression Line')

plt.scatter(x , y , color ='#ef5423' , label = 'Scatter Plot')

plt.xlabel("Head Size in cm3")
plt.ylabel("Brain Weight in gram")
plt.legend()
plt.show()
```

```python
ss_t = 0
ss_r = 0

for i in range(n):
    y_pred = d0 + b1 * x[i]
    ss_t += (y[i] - mean_y) **2
    ss_r += (y[i] - y_pred) ** 2
r2 = 1 - (ss_r/ss_t)

print(r2)
```

```python
# Using sklearn Library to solve Regression Problem

from  sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

x = x.reshape(1000, 1)

reg = LinearRegression()

reg = reg.fit(x,y)

y_pred = reg.predict(x)

mse = mean_squared_error(y,y_pred)

rmsg = np.sqrt(mse)

r2_score = reg.score(x , y)

print(np.sqrt(mse))
print(r2_score)
```
